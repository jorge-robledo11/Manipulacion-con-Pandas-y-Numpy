```python
import pandas as pd

csv = pd.read_csv('bestsellers-with-categories.csv')
csv
```

```python
# Groupby: Agrupa el Dataframe basado en un feature y espera un método adicional
csv.groupby('Genre').count().reset_index()
```

```python
csv.groupby('Genre').mean().reset_index()
```

```python
csv.groupby('Year').mean()
```

```python
csv.groupby('Year').std().reset_index()
```

```python
csv.groupby('Year').median()
```

```python
csv.groupby('Author').max()
```

```python
# Agrupar los datos por año y hacerle una sumatoria
csv.groupby('Year').sum().reset_index()
```

```python
csv.groupby('Year').max().reset_index()
```

## Agregaciones

```python
csv.groupby('Author').agg({'Reviews': ['min', 'max'], 'User Rating': 'sum'}).reset_index()
```

```python
csv.groupby(['Author','Year']).count()
```
