```python
import pandas as pd
import numpy as np
```

```python
dicc = {
    'Col 1': [1,2,3,np.nan,5,np.nan,7],
    'COl 2': list(range(7)),
    'COl 3': ['a', np.nan, 'c', 'd', None, 'f', 'g']
}

df = pd.DataFrame(dicc)
data = df.copy()
data.head(10)
```

```python
data.isnull()
```

```python
# 1 = Valores nulos | 0 = No
data.isnull() * 1
```

```python
data = df.fillna('Missing')
data.head(10)
```

```python
data = df.fillna(df.mean())
data.head(10)
```

```python
data = df.fillna(df.median())
data.head(10)
```

```python
data = df.fillna(df.mode())
data.head(10)
```

```python
data = df.interpolate()
data.head(10)
```

```python
df.dropna(inplace=True)
df.head(10)
```
