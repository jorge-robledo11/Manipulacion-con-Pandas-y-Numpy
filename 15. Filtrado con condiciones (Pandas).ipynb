```python
import pandas as pd
```

```python
csv = pd.read_csv('bestsellers-with-categories.csv')
csv
```

```python
# Filtrando por años mayores a un año
mayor_a_2016 = csv['Year'] > 2016
csv[mayor_a_2016]
```

```python
# Filtrando por un género y por años mayores a 2016
fiction = csv['Genre'] == 'Fiction'
csv[fiction & mayor_a_2016]
```

```python
# Filtrando negando un género y por años menores a 2016 incluyéndolo
csv[~fiction & ~mayor_a_2016]
```
