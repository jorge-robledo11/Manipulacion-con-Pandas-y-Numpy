```python
csv = pd.read_csv('bestsellers-with-categories.csv')
csv
```

```python
# Función
def two_times(value):
    return value * 2
```

```python
csv[['User Rating']].apply(two_times)
```

```python
csv['Rating 2n'] = csv[['User Rating']].apply(two_times)
csv
```

```python
# Aplicando funciones más avanzadas
csv['Rating 2^n'] = csv.apply(lambda x: x['User Rating'] ** 2 if x['Genre'] == 'Fiction' else x['User Rating'], axis=1)
csv
```
