```python
import pandas as pd

csv = pd.read_csv('bestsellers-with-categories.csv')
csv.head()
```

## Eliminar

```python
# Eliminar columnas
csv_copy = csv.copy()
csv.drop('Year', axis=1, inplace=True)
csv.head()
```

```python
# Eliminar una columna de otro modo
del csv['Genre']
csv.head()
```

```python
csv_copy.head()
```

```python
csv_copy.drop(0, axis=0).head()
```

```python
csv_copy.drop([0,1,2,3,4,5], axis=0).head()
```

```python
# Eliminar índices en rangos
csv_copy.drop(range(10), axis=0).head()
```

## Agregar

```python
# Agregar una columna con valores nan
import numpy as np
csv['Nueva Columna'] = np.nan
csv.head()
```

```python
# Concatenar una columna en un Dataframe
csv = pd.concat((csv, csv_copy['Genre']), axis=1)
csv.head()
```

```python
rango = range(0, csv.shape[0])
print(rango)
```

```python
# Crear una columna basado en un rango de valores
csv['Rango'] = rango
csv.head(10)
```

```python
# Concatenar por filas = Colocar un Dataframe sobre otro Dataframe
pd.concat((csv, csv), axis=0, ignore_index=True)
```
