```python
import pandas as pd
import numpy as np
```

## Concat

```python
datos =  {
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}

data = pd.DataFrame(datos)
data
```

```python
datos2 =  {
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
}

data2 = pd.DataFrame(datos2)
data2
```

```python
concatenar = pd.concat((data, data2), ignore_index=True)
concatenar
```

```python
pd.concat((data, data2), axis=1)
```

## Merge

```python
izq = {
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}

der = {
    'key': ['k0', 'k1', 'k2', 'k3'],
    'E': ['E0', 'E1', 'E2', 'E3'],
    'F': ['F0', 'F1', 'F2', 'F3'],
    'G': ['G0', 'G1', 'G2', 'G3'],
    'H': ['H0', 'H1', 'H2', 'H3']
}
data3 = pd.DataFrame(izq)
data4 = pd.DataFrame(der)
```

```python
data3
```

```python
data4
```

```python
# Merge
pd.merge(data3, data4, on='key')
```

```python
izq = {
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}

der = {
    'key_2': ['k0', 'k1', 'k2', 'k3'],
    'E': ['E0', 'E1', 'E2', 'E3'],
    'F': ['F0', 'F1', 'F2', 'F3'],
    'G': ['G0', 'G1', 'G2', 'G3'],
    'H': ['H0', 'H1', 'H2', 'H3']
}
data3 = pd.DataFrame(izq)
data4 = pd.DataFrame(der)
```

```python
# Merge
pd.merge(data3, data4, left_on='key', right_on='key_2')
```

```python
data3.merge(data4, left_on='key', right_on='key_2', how='right')
```
